```python
#Question (a)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('expectancy.csv')

#------------------------------------------
#Question (b)-
# Extract the columns from the DataFrame
year = data['Year']
age = data['Age']

# Perform linear regression (OLS)
slope, intercept = np.polyfit(year, age, 1)


# Generate the OLS line
ols_line = slope * np.array(year) + intercept

# Create the scatter plot with OLS line
plt.scatter(year, age, label='Age at Each Year')
plt.plot(year, ols_line, color='red', label='OLS Line')
plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Hong Kong Life(Age Over Time)')
plt.legend()

# Display the plot
plt.show()
print('slope ')
print(slope)
print('intercept ')
print(intercept)
```

```python
#question (C)
"""
Q(C)
To estimate the life expectancy for the year 2025 using the
provided data, i use the linear regression equation obtained
from the OLS line. The general formula for a straight line is:

y = mx + b
where:
y is the dependent variable (life expectancy)
x is the independent variable (year)
m is the slope of the line
b is the y-intercept of the line
In this case, the linear regression equation can be expressed as:
life_expectancy = slope * year + intercept

To calculate the life expectancy for the year 2025,  substitute the
value of 2025 for the year variable:

life_expectancy_2025 = slope * 2025 + intercept
Using the previously calculated slope and intercept,
i can calculate the life expectancy for 2025.

"""

# Calculate life expectancy for the year 2025
year_2025 = 2025
life_expectancy_2025 = slope * year_2025 + intercept

life_expectancy_2025=format(life_expectancy_2025, ".6f")
print("Life expectancy in 2025:", life_expectancy_2025)
#print('slope ', slope)
#print('intercept  ',intercept)
#print('year 2025  ', year_2025)
```

```python
#question (d)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
data = pd.read_csv('expectancy.csv')

# Extract the columns from the DataFrame
year = data['Year']
age = data['Age']

# Perform linear regression (OLS)
slope, intercept = np.polyfit(year, age, 1)
ols_line = slope * np.array(year) + intercept

# Create the scatter plot using Seaborn
sns.scatterplot(data=data, x='Year', y='Age', label='Age Over Time')

# Plot the OLS line
plt.plot(year, ols_line, color='red', label='OLS Line')

# Set the plot labels and title
plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Hong Kong Life(Age Over Time)')

# Display the plot
plt.legend()
plt.show()

```


Question(G)

In regression analysis, the coefficient of determination, often denoted as R-squared (R²),
is a statistical measure that  represents the proportion of the variance  in the 
dependent variable  that can be explained by the independent variables in a regression 
model.

It provides an indication of how well the regression line or model fits the observed data.

R-squared typically ranges from 0 to 1. A value of 0 indicates that the independent
variables explain none of the variance in the dependent variable, while a value of 
1 indicates that the independent variables explain all of the variance. However,
it's important to note that an R-squared value of 1 doesn't necessarily mean the
model is perfect or accurate; it simply means that the model can account for all
the observed variation.

Interpreting R-squared can be subjective and context-dependent. A higher R-squared 
value generally suggests a better fit of the regression line to the data, indicating that a
larger proportion of the variability in the dependent variable is explained by the
independent variables. On the other hand, a lower R-squared value implies that the
regression line does not capture much of the variability in the data and may not be a
good fit.

It's worth noting that R-squared should not be the sole criterion for evaluating the 
quality of a regression model. It is important to consider other factors such as the
significance of individual coefficients, the presence of multicollinearity, the overall
model fit, and the appropriateness of the model for the specific data and research question.

In summary, R-squared provides a measure of how well the regression line 
represents the variation in the dependent variable. However, 
it is crucial to consider the limitations of R-squared and 
other statistical measures and examine the overall model fit, 
significance of predictors, and potential confounding factors to draw
 meaningful conclusions from the regression analysis.
